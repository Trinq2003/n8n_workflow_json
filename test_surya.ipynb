```python
# OCR
from PIL import Image
from PIL import ImageDraw
from surya.recognition import RecognitionPredictor
from surya.detection import DetectionPredictor
from PIL import Image
import numpy as np
file = 'ocr_vi.png'
image_path = f'./image/{file}'
min_gray_input = 200
max_gray_input = 230

def set_gray_pixels_in_range_to_white(image_path, min_gray, max_gray):
    try:
        if not (0 <= min_gray <= max_gray <= 255):
            raise ValueError("Gray range must be between 0 and 255, with min_gray <= max_gray.")
        
        image = Image.open(image_path).convert('RGB')
        image_array = np.array(image)
        gray_mask = (image_array[:, :, 0] == image_array[:, :, 1]) & \
                    (image_array[:, :, 1] == image_array[:, :, 2])
        
        gray_values = image_array[:, :, 0]  # R channel is sufficient since R = G = B
        range_mask = (gray_values >= min_gray) & (gray_values <= max_gray)
        
        final_mask = gray_mask & range_mask
        image_array[final_mask] = [255, 255, 255]
        
        modified_image = Image.fromarray(image_array)
        modified_image.save(image_path)
        
        print(f"Image saved with gray pixels between {min_gray} and {max_gray} set to white at '{image_path}'")
        
    except FileNotFoundError:
        print(f"Error: Image file '{image_path}' not found.")
    except ValueError as e:
        print(f"Input error: {str(e)}")
    except Exception as e:
        print(f"Error: {str(e)}")


try:
    # Process the image
    set_gray_pixels_in_range_to_white(image_path, min_gray_input, max_gray_input)
    
except ValueError as e:
    print(f"Input error: {str(e)}")
except Exception as e:
    print(f"Error: {str(e)}")

image = Image.open(image_path)
langs = ["vi"] # Replace with your languages or pass None (recommended to use None)
recognition_predictor = RecognitionPredictor()
detection_predictor = DetectionPredictor()
predictions = recognition_predictor([image], [langs], detection_predictor)

image_copy = image.copy().convert('RGB')
draw = ImageDraw.Draw(image_copy)

text_lines = predictions[0].text_lines

# Draw bounding boxes
for idx, line in enumerate(text_lines):
    bbox = line.bbox
    if bbox:
        print(f"{idx}: {line.text}")
        
        x1, y1, x2, y2 = bbox
        draw.rectangle((x1, y1, x2, y2), outline='red', width=2)
        draw.text((x1, y1), f'{idx}', fill='blue', size=300)

# Save the image with bounding boxes
image_copy.save(f'image/results/{file}')
```

```python
# Layout & reading order
from PIL import Image
from PIL import ImageDraw
from surya.layout import LayoutPredictor

file = 'layout_vi.png'
image_path = f'./image/{file}'

image = Image.open(image_path)
layout_predictor = LayoutPredictor()

image_copy = image.copy().convert('RGB')
draw = ImageDraw.Draw(image_copy)

# layout_predictions is a list of dicts, one per image
layout_predictions = layout_predictor([image])

for idx, bbox in enumerate(layout_predictions[0].bboxes):
    label = bbox.label
    reading_order = bbox.position
    x1, y1, x2, y2 = bbox.bbox
    draw.rectangle((x1, y1, x2, y2), outline='red', width=2)
    draw.text((x1, y1), f'{idx}. {label} [{reading_order}]', fill='blue', size=300)
    
image_copy.save(f'image/results/{file}')
```

```python
# Table recognition
from PIL import Image
from PIL import ImageDraw
import numpy as np
from surya.table_rec import TableRecPredictor
from surya.layout import LayoutPredictor

file = 'table_vi.png'
image_path = f'./image/{file}'

image = Image.open(image_path)
table_rec_predictor = TableRecPredictor()

layout_predictor = LayoutPredictor()
layout_predictions = layout_predictor([image])

tables = []
for idx, bbox in enumerate(layout_predictions[0].bboxes):
    label = bbox.label
    if label.lower() == 'table':
        x1, y1, x2, y2 = bbox.bbox
        table_image = Image.fromarray(np.array(image)[int(y1):int(y2), int(x1):int(x2)])
        tables.append(table_image)

print(f"[INFO] Recognized {len(tables)} tables")

for table in tables:
    draw = ImageDraw.Draw(table)
    table_predictions = table_rec_predictor([table])

    for idx, cell in enumerate(table_predictions[0].cells):
        is_header = cell.is_header
        cell_id = cell.cell_id
        row_id = cell.row_id
        col_id = cell.col_id
        x1, y1, x2, y2 = cell.bbox
        
        if is_header:
            draw.rectangle((x1, y1, x2, y2), outline='red', width=2)
            draw.text((x1, y1), f'{idx}. {cell_id} [{row_id}, {col_id}]', fill='blue', size=300)
        else:
            draw.rectangle((x1, y1, x2, y2), outline='green', width=2)
            draw.text((x1, y1), f'{idx}. {cell_id} [{row_id}, {col_id}]', fill='blue', size=300)

    table.save(f'image/results/table/table_{idx}.png')
```

```python
from pdf2image import convert_from_path
from PIL import Image
import numpy as np

def set_gray_pixels_in_range_to_white(image, min_gray, max_gray):
    """
    Sets gray pixels within a specified range to white in the given image.
    
    Args:
        image: PIL Image object
        min_gray (int): Minimum gray value (0-255)
        max_gray (int): Maximum gray value (0-255)
    
    Returns:
        PIL Image object with modified pixels, or the original image if an error occurs
    """
    try:
        if not (0 <= min_gray <= max_gray <= 255):
            raise ValueError("Gray range must be between 0 and 255, with min_gray <= max_gray.")
        
        image = image.convert('RGB')
        image_array = np.array(image)
        gray_mask = (image_array[:, :, 0] == image_array[:, :, 1]) & \
                    (image_array[:, :, 1] == image_array[:, :, 2])
        
        gray_values = image_array[:, :, 0]
        range_mask = (gray_values >= min_gray) & (gray_values <= max_gray)
        
        final_mask = gray_mask & range_mask
        image_array[final_mask] = [255, 255, 255]
        
        modified_image = Image.fromarray(image_array)
        return modified_image
        
    except ValueError as e:
        print(f"Input error: {str(e)}")
        return image
    except Exception as e:
        print(f"Error processing image: {str(e)}")
        return image

def process_pdf(pdf_path, min_gray, max_gray):
    """
    Processes a PDF by converting pages to images, modifying gray pixels, and saving back to PDF.
    
    Args:
        pdf_path (str): Path to the input PDF file
        min_gray (int): Minimum gray value (0-255)
        max_gray (int): Maximum gray value (0-255)
    """
    try:
        # Validate gray range once before processing
        if not (0 <= min_gray <= max_gray <= 255):
            raise ValueError("Gray range must be between 0 and 255, with min_gray <= max_gray.")
        
        # Convert PDF pages to a list of PIL Image objects
        pages = convert_from_path(pdf_path)
        
        # Process each page
        modified_pages = []
        for i, page in enumerate(pages):
            print(f"Processing page {i + 1} of {len(pages)}")
            modified_page = set_gray_pixels_in_range_to_white(page, min_gray, max_gray)
            modified_pages.append(modified_page)
        
        # Save all modified pages back to the original PDF path
        if modified_pages:
            modified_pages[0].save(
                pdf_path,
                save_all=True,
                append_images=modified_pages[1:]
            )
            print(f"Processed PDF saved at '{pdf_path}'")
        else:
            print("No pages were processed.")
            
    except ValueError as e:
        print(f"Input error: {str(e)}")
    except FileNotFoundError:
        print(f"Error: PDF file '{pdf_path}' not found.")
    except Exception as e:
        print(f"Error during PDF processing: {str(e)}")

# Example usage
if __name__ == "__main__":
    pdf_file = "example.pdf"
    process_pdf(pdf_file, min_gray=100, max_gray=150)
```
